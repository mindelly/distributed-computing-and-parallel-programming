```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
threads = np.array([12.523, 6.488, 4.310, 4.085, 3.934])
openmp = np.array([13.980, 7.400, 4.516, 4.735, 4.434])
p = np.array([1, 2, 4, 6, 8])
```

```python
fig,ax=plt.subplots(1,1,figsize=(20,10))
ax.grid(b=True, axis='both')
ax.plot(p, threads, 'o-', label='POSIX')
ax.plot(p, openmp, 'o-', color='red', label='OpenMP')
ax.set_xlabel(' ', fontsize=25)
ax.set_ylabel(' , ', fontsize=25)
plt.tick_params(axis='both', which='major', labelsize=25)
ax.legend(prop={'size': 25})
```

```python
tp = 13.622
sth = tp / threads
sop = tp / openmp
np.set_printoptions(precision=3)
print(sth, '\n', sop)
```

```python
fig,ax=plt.subplots(1,1,figsize=(20,10))
ax.grid(b=True, axis='both')
ax.plot(p, sth, 'o-', label='POSIX')
ax.plot(p, sop, 'o-', color='red', label='OpenMP')
ax.set_xlabel(' ', fontsize=25)
ax.set_ylabel('', fontsize=25)
plt.tick_params(axis='both', which='major', labelsize=25)
ax.legend(prop={'size': 25})
```

```python
t1 = threads[0]
o1 = openmp[0]
srth = t1 / threads
srop = o1 / openmp
np.set_printoptions(precision=3)
print(srth, '\n', srop)
```

```python
fig,ax=plt.subplots(1,1,figsize=(20,10))
ax.grid(b=True, axis='both')
ax.plot(p, srth, 'o-', label='POSIX')
ax.plot(p, srop, 'o-', color='red', label='OpenMP')
ax.set_xlabel(' ', fontsize=25)
ax.set_ylabel(' ', fontsize=25)
plt.tick_params(axis='both', which='major', labelsize=25)
ax.legend(prop={'size': 25})
```

```python
eth = sth / p
eop = sop / p
print(eth, '\n', eop)
```

```python
fig,ax=plt.subplots(1,1,figsize=(20,10))
ax.grid(b=True, axis='both')
ax.plot(p, eth, 'o-', label='POSIX')
ax.plot(p, eop, 'o-', color='red', label='OpenMP')
ax.set_xlabel(' ', fontsize=25)
ax.set_ylabel('', fontsize=25)
plt.tick_params(axis='both', which='major', labelsize=25)
ax.legend(prop={'size': 25})
```
